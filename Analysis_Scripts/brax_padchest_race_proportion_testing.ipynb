# BRAX is a dataset taken from the country of Brazil. In this country of ~203 million individuals, people are asked to identify as White (43.5%), Black (10.2%), Yellow (0.4%), Brown (45.3%) or Indigenous (0.6%) in the country's census.

We compare the proportion of individuals identifying as 'White' in the Brazilian national census with the proportion of individuals predicted be 'White' in BRAX dataset by our MIMIC and CheXpert trained models using a one-sample proportion z-test.

**MIMIC**

```python
from statsmodels.stats.proportion import proportions_ztest


count = 15498   # number of predicted 'White' individuals
n_obs = 19309            # total observations
p_0 = 0.434566636              # census proportion

# Run one-sample proportion z-test
z_stat, p_val = proportions_ztest(count, n_obs, value=p_0, alternative='two-sided')

print(f"Z statistic: {z_stat:.3f}")
print(f"P-value: {p_val:.3e}")
```

**CheXpert**

```python
from statsmodels.stats.proportion import proportions_ztest


count = 13625   # number of predicted 'White' individuals
n_obs = 19309            # total observations
p_0 = 0.434566636              # census proportion

# Run one-sample proportion z-test
z_stat, p_val = proportions_ztest(count, n_obs, value=p_0, alternative='two-sided')

print(f"Z statistic: {z_stat:.3f}")
print(f"P-value: {p_val:.3e}")
```
