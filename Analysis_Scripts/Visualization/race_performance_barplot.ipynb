```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data = [
    ("MIMIC", "Asian", 0.9246, [0.9187, 0.9311]),
    ("MIMIC", "Black", 0.9487, [0.9462, 0.9512]),
    ("MIMIC", "Hispanic/Latino", 0.7892, [0.7799, 0.7978]),
    ("MIMIC", "White", 0.9128, [0.9095, 0.9160]),
    ("CheXpert", "Asian", 0.8565, [0.8506, 0.8631]),
    ("CheXpert", "Black", 0.9222, [0.9152, 0.9290]),
    ("CheXpert", "Hispanic/Latino", 0.7098, [0.7014, 0.7179]),
    ("CheXpert", "White", 0.8349, [0.8296, 0.8396]),
]

# Plot config
datasets = ["MIMIC", "CheXpert"]
race_labels = ["Asian", "Black", "Hispanic/Latino", "White"]
purplecolor = '#9b59b6'
colors = {
    "Asian": 'skyblue',
    "Black": 'lightgreen',
    "Hispanic/Latino": 'salmon',
    "White": purplecolor,
}

bar_width = 0.25
x = np.arange(len(datasets))

group_gap = 0.5  # spacing between MIMIC and CheXpert
group_centers = [0, 1 + group_gap]  # center positions for MIMIC and CheXpert

# Create the plot
fig, ax = plt.subplots(figsize=(9, 6))

# Plot each race as a group across MIMIC and CheXpert
for i, race in enumerate(race_labels):
    means, lower_err, upper_err = [], [], []

    for dataset in datasets:
        match = next((d for d in data if d[0] == dataset and d[1] == race), None)
        if match:
            mean = match[2]
            ci = match[3]
            means.append(mean)
            lower_err.append(mean - ci[0])
            upper_err.append(ci[1] - mean)

    # Position for each bar within the group
    for j, center in enumerate(group_centers):
        offset = (i - 1.5) * bar_width
        ax.bar(center + offset, means[j], width=bar_width,
               yerr=[[lower_err[j]], [upper_err[j]]], capsize=4,
               color=colors[race], edgecolor='black', label=race if j == 0 else None)

# Formatting
ax.set_xticks(group_centers)
ax.set_xticklabels(datasets, fontsize=16)
ax.set_ylabel("AUROC", fontsize=14, fontweight='bold')
ax.set_title("AUROC by Race and Dataset", fontsize=15, fontweight='bold')
ax.set_ylim(0.6, 1.0)
ax.grid(axis='y', linestyle='--', alpha=0.6)
#ax.legend(title="Race", fontsize=11, title_fontsize=12)
plt.tight_layout()
plt.show()
```
