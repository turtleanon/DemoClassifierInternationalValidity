```python
test_df = pd.read_csv('/content/s3/PadChest_Full_Dataset/Docs/PADCHEST_chest_x_ray_images_labels_160K_01.02.19.csv')
test_df['Path'] = '/content/s3/PadChest_Full_Dataset/' + test_df['ImageDir'].astype(str) + '/' + test_df['ImageID'].astype(str)
```

```python
test_df1 = test_df[0:53620]
test_df2 = test_df[53620:107240]
test_df3 = test_df[107240:160860]
```

```python
count = 0
for path in test_df1['Path']:
  count = count + 1
  if count%10000 == 0:
    print(count)
  try:
            img = Image.open(path) # open the image file
            img.verify() # verify that it is, in fact an image
  except (IOError, SyntaxError) as e:
            print('Bad file:', path) # print out the names of corrupt files


```
