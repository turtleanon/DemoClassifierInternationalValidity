```python
import pandas as pd
import pydicom as dicom
import os
import cv2
import numpy as np
from PIL import Image, ImageOps
import pylibjpeg
import matplotlib.pyplot as plt
```

```python
folder_path = '/content/s3/TBXPredict/data/Dataset_B'
jpg_folder_path = '/content/India_TB'
train_images_path = os.listdir(os.path.join(folder_path, 'Training_Data'))[1:]
test_images_path = os.listdir(os.path.join(folder_path, 'Testing_Data'))[1:]
paths = []
```

```python
for n, image in enumerate(train_images_path):
    ds = dicom.dcmread(os.path.join(folder_path, 'Training_Data', image))
    new_image = ds.pixel_array.astype(float)
    scaled_image = (np.maximum(new_image, 0) / new_image.max()) * 255.0
    scaled_image = np.uint8(scaled_image)
    final_image = Image.fromarray(scaled_image)
    image = image.replace('.dcm', '.jpg')
    final_image.save(os.path.join(jpg_folder_path, image))
    paths.append(image)
for n, image in enumerate(test_images_path):
    ds = dicom.dcmread(os.path.join(folder_path, 'Testing_Data', image))
    new_image = ds.pixel_array.astype(float)
    scaled_image = (np.maximum(new_image, 0) / new_image.max()) * 255.0
    scaled_image = np.uint8(scaled_image)
    final_image = Image.fromarray(scaled_image)
    image = image.replace('.dcm', '.jpg')
    final_image.save(os.path.join(jpg_folder_path, image))
    paths.append(image)
```
